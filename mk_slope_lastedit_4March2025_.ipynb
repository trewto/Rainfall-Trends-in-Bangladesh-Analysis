```python
import pandas as pd
import pymannkendall as mk

# Function to calculate Mann-Kendall slope (Sen's slope) and Zmk for each month
def calculate_mk_slope(df,city_name):
    # Initialize a result dictionary to store Zmk and Q values for each month
    results = {}

    # Loop over each month (1 to 12)
    for month in range(1, 13):

        # Filter data for the specific month
        month_data = df[df['Month'] == month]
        print(month_data)
        # Perform Mann-Kendall test (it returns Zmk and Sen's slope Q)
        mk_test_result = mk.original_test(month_data['sum_1_to_31'])
        print(mk_test_result)
        # Store the results in the dictionary
        #results[month] = {'City':city_name,'Zmk': mk_test_result.z, 'Q (Sen\'s slope)': mk_test_result.slope}
        #for main
        #results[month] = {f"Zmk_{city_name}": mk_test_result.z, f'Q (Sen\'s slope)': mk_test_result.slope}
        #for only trend
        results[month] = {f"Zmk_{city_name}": mk_test_result.trend}

    # Convert the result dictionary into a DataFrame for better visualization
    results_df = pd.DataFrame(results)
#    results_df = pd.DataFrame(results).T
    return results_df

```

```python

# Create an empty list to store statistics
table_data = []

for ind in preferred_index:
    city = unique_city[ind]
    df_city = df[df['Stati'] == city]

    for i in range(1, 13):  # Loop through months

        df_month = df_city[df_city['Month'] == i]
        print(df_month)

        if not df_month.empty:  # Only process if data exists for the month
            stats = df_month['sum_1_to_31'].describe()
            mean_val = df_month['sum_1_to_31'].mean()
            std_val = df_month['sum_1_to_31'].std()
            var_val = df_month['sum_1_to_31'].var()

            table_data.append([city, i, stats['count'], stats['mean'], stats['std'], stats['min'], stats['25%'], stats['50%'], stats['75%'], stats['max'], mean_val, std_val, var_val])
        else:
            print(f"No data found for {city} in month {i}")
# Convert to DataFrame for better visualization
stats_df = pd.DataFrame(table_data, columns=['City', 'Month', 'Count', 'Mean', 'Std Dev', 'Min', '25%', 'Median', '75%', 'Max', 'Mean_2', 'Std_2', 'Var'])

# Print table in a structured format
print(tabulate(stats_df, headers='keys', tablefmt='pretty'))
```

```python
# prompt: now i want to sum all the month of the year, that mean i want yearly rainfall, only show the yearly rainfall df

# Yearly rainfall
yearly_rainfall = df.groupby(['Stati', 'Year'])['sum_1_to_31'].sum().reset_index()
yearly_rainfall = yearly_rainfall.rename(columns={'sum_1_to_31': 'Yearly_Rainfall'})
yearly_rainfall
```

```python

# Create an empty list to store statistics
table_data_yearly = []

for ind in preferred_index:
    city = unique_city[ind]
    df_city_yearly = yearly_rainfall[yearly_rainfall['Stati'] == city]

    if not df_city_yearly.empty:  # Only process if data exists for the month
        stats = df_city_yearly['Yearly_Rainfall'].describe()
        mean_val = df_city_yearly['Yearly_Rainfall'].mean()
        std_val = df_city_yearly['Yearly_Rainfall'].std()
        var_val = df_city_yearly['Yearly_Rainfall'].var()

        table_data_yearly.append([city, stats['count'], stats['mean'], stats['std'], stats['min'], stats['25%'], stats['50%'], stats['75%'], stats['max'], mean_val, std_val, var_val])
    else:
        print(f"No data found for {city} ")

# Convert to DataFrame for better visualization
stats_df_yearly = pd.DataFrame(table_data_yearly, columns=['City', 'Count', 'Mean', 'Std Dev', 'Min', '25%', 'Median', '75%', 'Max', 'Mean_2', 'Std_2', 'Var'])

# Print table in a structured format
print(tabulate(stats_df_yearly, headers='keys', tablefmt='pretty'))
```
