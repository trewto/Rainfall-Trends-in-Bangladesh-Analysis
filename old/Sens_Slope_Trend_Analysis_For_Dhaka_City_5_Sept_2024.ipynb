```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def sens_slope(x, y):
    """Calculate Sen's slope estimator"""
    n = len(x)
    slopes = []
    for i in range(n):
        for j in range(i + 1, n):
            if x[i] != x[j]:
                slopes.append((y[j] - y[i]) / (x[j] - x[i]))
    return np.median(slopes)

# Example DataFrame (replace with your actual DataFrame)
# df_dhaka = pd.read_csv('your_file.csv')

# Ensure correct data types
df_dhaka['Year'] = df_dhaka['Year'].astype(int)
df_dhaka['Month'] = df_dhaka['Month'].astype(int)

# Create a Date column for plotting
df_dhaka['Date'] = pd.to_datetime(df_dhaka[['Year', 'Month']].assign(DAY=1))

# Sort by Date to ensure proper trend calculation
df_dhaka = df_dhaka.sort_values('Date')

# Prepare data for Sen's Slope calculation
dates = df_dhaka['Date']
rainfall = df_dhaka['sum_1_to_31']
x = np.arange(len(dates))  # Numeric representation of time

# Calculate Sen's Slope
slope = sens_slope(x, rainfall)
print(f"Sen's Slope: {slope}")

# Plotting the trend
plt.figure(figsize=(12, 6))
plt.plot(dates, rainfall, 'o', label='Monthly Rainfall')
plt.plot(dates, slope * x + rainfall.mean(), 'r-', label='Sen\'s Slope Trend')
plt.xlabel('Date')
plt.ylabel('Monthly Rainfall')
plt.title('Monthly Rainfall Trend for Dhaka with Sen\'s Slope')
plt.legend()
plt.grid(True)
plt.show()
```

```python

# Define the Sen's slope function
def sens_slope(x, y):
    """Calculate Sen's slope estimator"""
    n = len(x)
    slopes = []
    for i in range(n):
        for j in range(i + 1, n):
            if x[i] != x[j]:
                slopes.append((y[j] - y[i]) / (x[j] - x[i]))
    return np.median(slopes)
```

```python
# Ensure 'Month' column is numeric
df_dhaka['Month'] = pd.to_numeric(df_dhaka['Month'], errors='coerce')

# Handle missing values in 'Month'
df_dhaka = df_dhaka.dropna(subset=['Month'])

# Print DataFrame after conversions
print("\nDataFrame after handling missing values:")
print(df_dhaka.head())

# Create a Date column for sorting
df_dhaka['Date'] = pd.to_datetime(df_dhaka[['Year', 'Month']].assign(DAY=1))
df_dhaka = df_dhaka.sort_values('Date')

# Print DataFrame after sorting
print("\nDataFrame after sorting:")
print(df_dhaka.head())
```

```python

# Initialize dictionary to hold trends for each month
trends = {}

# Analyze each month
for month in range(1, 13):
    # Filter data for the current month
    month_data = df_dhaka[df_dhaka['Month'] == month]

    if month_data.empty:
        print(f"No data available for month {month}.")
        continue

    # Prepare data for Sen's Slope calculation
    years = month_data['Year'].values
    rainfall = month_data['sum_1_to_31'].values
    x = np.arange(len(years))  # Numeric representation of time

    # Calculate Sen's Slope
    slope = sens_slope(x, rainfall)
    trends[month] = slope

    # Plotting the trend for each month
    plt.figure(figsize=(10, 6))
    plt.plot(years, rainfall, 'o', label=f'Month {month} Rainfall')
    plt.plot(years, slope * x + np.mean(rainfall), 'r-', label=f'Sen\'s Slope Trend for Month {month}')
    plt.xlabel('Year')
    plt.ylabel('Rainfall')
    plt.title(f'Rainfall Trend for Month {month}')
    plt.legend()
    plt.grid(True)
    plt.show()

# Display trends for each month
for month, slope in trends.items():
    print(f"Sen's Slope for Month {month}: {slope}")
```
