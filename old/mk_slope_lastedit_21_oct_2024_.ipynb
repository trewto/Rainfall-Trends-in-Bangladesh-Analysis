```python
import pandas as pd
import pymannkendall as mk

# Function to calculate Mann-Kendall slope (Sen's slope) and Zmk for each month
def calculate_mk_slope(df):
    # Initialize a result dictionary to store Zmk and Q values for each month
    results = {}

    # Loop over each month (1 to 12)
    for month in range(1, 13):
        # Filter data for the specific month
        month_data = df[df['Month'] == month]

        # Perform Mann-Kendall test (it returns Zmk and Sen's slope Q)
        mk_test_result = mk.original_test(month_data['sum_1_to_31'])

        # Store the results in the dictionary
        results[month] = {'Zmk': mk_test_result.z, 'Q (Sen\'s slope)': mk_test_result.slope}

    # Convert the result dictionary into a DataFrame for better visualization
    results_df = pd.DataFrame(results).T
    return results_df

```

```python
# dataframe is named df_city
mk_results = calculate_mk_slope(df_city)
print(mk_results)
```

```python
for city in unique_city:
    if city in ['          Daily  total  Rainfall  data  in  Millimeter  ', 'Stati', np.nan]:
        continue

    # Filter the data for the specific city
    df_city = df[df['Stati'] == city]
    df_city = df_city[['Year', 'Month', 'sum_1_to_31']]

    # Check if the filtered dataframe has enough data points and unique values
    if len(df_city) > 2 and df_city['sum_1_to_31'].nunique() > 1:  # At least 2 data points and multiple unique values
        try:
            mk_results = calculate_mk_slope(df_city)
            print(city)
            print(mk_results)
        except ZeroDivisionError as e:
            print(f"Error calculating MK slope for {city}: {e}")
    else:
        print(f"Not enough data or unique values for {city}")
```
